```python
def histograms_plot(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax)
        ax.set_title(feature+" Distribution")

    fig.tight_layout()  
    plt.show()
```

```python
GR_min = df['GR'].min()
GR_max = df['GR'].max()
df['shale_volume'] = (df['GR'] - GR_min) / (GR_max - GR_min)
```

```python
df['Pore_Pressure'] = 0.052 * (df['Vp'] - df['Vp'].iloc[0]).cumsum() * (df['DEPT'] - z0)

# Visualize or analyze 'Pore_Pressure' column in 'data'
print(df['Pore_Pressure'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with columns 'Depth', 'RHOB', 'DT', 'GR', 'Pore_Pressure', 'Lithology'

data = df.values
mneumonics = ['RHOB', 'DT', 'GR', 'Pore_Pressure']
rows, cols = 1, 4

fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(8, 8), sharey=True)

colors = ['red', 'black', 'red']  # Define colors for line plots

for i in range(cols):
    if i < cols - 1:
        ax[i].plot(data[:, i + 1], data[:, 0], linewidth=0.5, color=colors[i])
        ax[i].set_ylim(max(data[:, 0]), min(data[:, 0]))
        ax[i].set_title(mneumonics[i])
        ax[i].minorticks_on()
        ax[i].grid(which='major', linestyle='-', linewidth=0.5, color='lime')
        ax[i].grid(which='minor', linestyle=':', linewidth=0.5, color='black')
    elif i == cols - 1:
        F = np.vstack((df['Pore_Pressure'], df['Pore_Pressure'])).T
        img = ax[i].imshow(F, aspect='auto', extent=[0, 1, max(data[:, 0]), min(data[:, 0])], cmap='viridis')  # Change colormap here
        ax[i].set_title(mneumonics[i])
        # Add colorbar for Pore Pressure
        cbar = plt.colorbar(img, ax=ax[i])
        cbar.set_label('Pore Pressure')
        cbar.ax.invert_yaxis()  # Invert colorbar for depth from lowest to highest

# GR Plot with fill_between
y2 = data[:, 2]  # GR
y1 = y2 * 0 + 70  # 70 API is Sand-Shale Boundary Line
ax[2].fill_betweenx(data[:, 0], y1, y2, where=(y2 >= y1), color='gold', linewidth=0)
ax[2].fill_betweenx(data[:, 0], y1, y2, where=(y2 < y1), color='lime', linewidth=0)
ax[2].set_xlim(0, 150)  # Set x-axis limit for GR log

plt.tight_layout()
plt.show()
```
